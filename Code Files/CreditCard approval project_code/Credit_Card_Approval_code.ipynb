```python
# create function for handling outliers

def handling_outliers(df , cols):

  # apply IQR on every column
  for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # define upper and lower bounds for outliers
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # replace upper and lower vlues with outliers
    df[col]  = np.where(df[col] > upper , upper,
                        np.where(df[col] < lower , lower , df[col]))

  return df


# names of columns
num_col = ["Income" , "Debt"]
df = handling_outliers(df , num_col)
```

```python
# import label encoder from sklearn.preprocessing

from sklearn.preprocessing import LabelEncoder

# LabelEncnder seve in variable

le = LabelEncoder()


# select attributes for encoding

cols = ["Gender" , "Married"]


# apply encoding on Attributes

for col in cols:

   df[col] = le.fit_transform(df[col])
```

```python
# appling one hot encodeing on attributes
one_hot_encoding = pd.get_dummies(df[["City" , "Education_Level"	 , "Employment_Type"	,"Housing_Status" , "Citizenship"]] , dtype=int)

# connent encoding columns with dataset
df = pd.concat([df , one_hot_encoding] , axis = 1)
```

```python
# spit data into x and y variable
x = df.drop("Approved" , axis = 1)

# y have target attribute = df["Target"]
y = df["Approved"]
```

```python
# check length of rows and columns
x.shape
```

```python
# we are import scalling
from sklearn.preprocessing import StandardScaler

# create an object
scaler = StandardScaler()

# scalling fit on data
x = scaler.fit_transform(x)
```

```python
# we are import train_test_split
from sklearn.model_selection import train_test_split

# splitting data into training and testing
X_train , X_test , Y_train , Y_test = train_test_split(x , y , test_size = 0.20 , random_state=42)
```

```python
# display len of rows and columns of train and test data

print("X_Train : " , X_train.shape)
print("Y_Train : " , Y_train.shape)
print("X_Test  : " ,X_test.shape)
print("Y_Test : ", Y_test.shape)
```

```python
# create a plot to check balance of 0 and 1 into target variable

df["Approved"].value_counts().plot(kind="bar" , color=["skyblue" , "orange"])
plt.title(".....Count of Approved (1) and Not Approved (0).....")
plt.xlabel(".....CLASS.....")
plt.ylabel("....COUNT....")
plt.show()
```

```python
# import DecisionTreeClassifire
from sklearn.tree import DecisionTreeClassifier

# Intialize DecisionTree model with hyperparameters
dt = DecisionTreeClassifier(criterion="gini",
                                  max_depth=10,
                                 class_weight="balanced",
                                  splitter="random")

# fit model on training data
dt.fit(X_train , Y_train)
```

```python
# model predict value frim test data
Y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report

# display accuracy_score
accuracy = accuracy_score(Y_test , Y_pred)
print(f"ACCURACY : {accuracy}")

# display classificatoin_report
print("\nCLASSIFICATION_REPORT")
print(classification_report(Y_test , Y_pred))

# display confusion_matrix
print("\nCONFUSION_MATRIX")
print(confusion_matrix(Y_test , Y_pred))
```
